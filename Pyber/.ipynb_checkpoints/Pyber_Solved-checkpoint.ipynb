```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data

city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)
```

```python
merge_table = pd.merge(ride_df, city_df, on="city", how="left")
merge_table.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types


grouped = merge_table.groupby(['city', 'type', 'driver_count'])

new_df = grouped['fare'].mean()
new_df = pd.DataFrame(new_df)
new_df['ride_count'] = pd.DataFrame(grouped['ride_id'].count())
new_df = new_df.rename(columns={'fare':'avg_fare'})
new_df = new_df.reset_index()
new_df.head()

```

```python
# Seperate "Urban", "Rural", "Suburban"

rural_data = new_df.loc[new_df['type'] == 'Rural']
# rural_data.head()
urban_data = new_df.loc[new_df['type'] == 'Urban']
# urban_data.head()
suburban_data = new_df.loc[new_df['type'] == 'Suburban']
# suburban_data.head()

# Build the scatter plots for each city types

fig, ax = plt.subplots(sharex=True)

rural_data.plot( kind='scatter', subplots=True, x='ride_count', y='avg_fare',
                   c='gold', ax=ax, alpha=0.5, edgecolors='black', label='RURAL', s=rural_data.driver_count*10)
urban_data.plot( kind='scatter', subplots=True, x='ride_count', y='avg_fare',
                   c='coral', ax=ax, alpha=0.5, edgecolors='black', label='URBAN', s=urban_data.driver_count*10)
suburban_data.plot( kind='scatter', subplots=True, x='ride_count', y='avg_fare',
                   c='lightskyblue', ax=ax, alpha=0.5, edgecolors='black', label='SUBURBAN', s=suburban_data.driver_count*10)

# Incorporate the other graph properties

ax.grid()
ax.set_axisbelow(False)
plt.xlabel("Total Number of Rides - Per City ")
plt.ylabel("Average Fare($)")
plt.title("Pyber Ride Sharing Data")

# Create a legend

lgnd = plt.legend(loc="upper right", title="City Types", fontsize="small")
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]

# Incorporate a text label regarding circle size

ax.text(45, 30, s="Note: \n Circle Size correlates with the driver count per city.")

# Save Figure

plt.savefig('Pyber_Scatterplot.png')
```

## Total Fares by City Type

```python
# Calculate Type Percents

tot_fares = merge_table.groupby(['type'])['fare'].sum()

# Build Pie Chart
labels = ["Rural", "Suburban", "Urban"]

plt.pie(tot_fares, explode=(0,0,.2), shadow=True, 
               colors=('lightskyblue','gold', 'lightcoral'), startangle=160, autopct="%1.1f%%", labels=labels)
plt.axis('tight')
plt.title("% of Total Fares by City Type",)

# Save Figure

plt.savefig('Pie_TotalFares_CityType')
```

## Total Rides by City Type

```python
# Calculate Ride Percents

tot_rides = merge_table.groupby(['type'])['ride_id'].count()
tot_rides.head()

# Build Pie Chart
labels = ["Rural", "Suburban", "Urban"]

plt.pie(tot_rides, explode=(0,0,.2), shadow=True, 
               colors=('lightskyblue','gold', 'lightcoral'), startangle=160, autopct="%1.1f%%", labels=labels)
plt.axis('tight')
plt.title("% of Total Rides by City Type",)

# Save Figure

plt.savefig('Pie_TotalRides_CityType')
```

## Total Drivers by City Type

```python
# Calculate Driver Percents

city_data = city_df.drop_duplicates('city', keep = 'first')
tot_drivers = city_data.groupby(['type'])['driver_count'].sum()


# Build Pie Chart
labels = ["Rural", "Suburban", "Urban"]

plt.pie(tot_drivers, explode=(0,0,.2), shadow=True, 
               colors=('lightskyblue','gold', 'lightcoral'), startangle=160, autopct="%1.1f%%", labels=labels)
plt.axis('tight')
plt.title("% of Total Drivers by City Type",)
# Save Figure

plt.savefig('Pie_TotalDrivers_CityType')
```

```python
## Observations:

## 1. Rural areas tend to have fewer rides per city in addition to higher fares. On the contrary, urban areas have lower average fares with more total rides.
## 2. Rural and Suburban areas have less than 20% of Pyber Drivers. 
## 3. Percentage of Total Rides corresponds closely to Percentage of Total Fares by City type. 
```
